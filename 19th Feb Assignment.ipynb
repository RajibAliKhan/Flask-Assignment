```python
App routing :
    In Flask, app routing is used as a decorator.
    It refers the process of binding / mapping the URLs to a function.
    By using this, it exposed to the outer world and accessable from anywhere.
    
Uses :
    1. From a single mother app route, we can access a huge number of URLs.
    2. Each route corresponds to a specific URL.
    3. App route allow to handle different HTTP methods.
    4. It provide a way to define the structure and behavior of web application.
```
